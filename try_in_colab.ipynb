```python
from src.data import download_av2_split, load_polars_dataframe, ArgoverseVehicleDataset
from src.train import train
from src.evaluate import visualize_good_predictions
import torch

DATA_DIR = "/content/argoverse2_data"
download_av2_split("val", 50, DATA_DIR)
print("Data downloaded!")
```

```python
print("Starting training on subset...")
model = train(DATA_DIR, epochs=1)
print("Training complete.")
```

```python
df_val = load_polars_dataframe(DATA_DIR)
if df_val is not None:
    dataset = ArgoverseVehicleDataset(df_val)
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    visualize_good_predictions(model, dataset, device, num_plots=5)
```
